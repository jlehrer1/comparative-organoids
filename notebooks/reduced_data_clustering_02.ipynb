## Reduced Data Clustering 02

In this notebook, we'll visualize the clusters generated by clustering over several different UMAP representations

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import plotly.express as px 
import seaborn as sns
```

First, let's define a method to visualize the clusters

```python
def plot(data, title=''):
    fig, ax = plt.subplots(figsize=(15, 10))
    g = sns.scatterplot(
        x='0', 
        y='1',
        data=data,
        ax=ax,
        s=1,
        palette='tab10',
    )
    
    g.set(xticklabels=[], yticklabels=[])
    plt.xlabel(None)
    plt.ylabel(None)
    
    plt.savefig('umap_cluster.png', dpi=300)
    plt.show()
    
def three_plot(data):
    fig = px.scatter_3d(data, x='0', y='1', z='2', color='label')
    fig.update_traces(marker=dict(size=1))
    fig.show()
```

Now we can read in the primary UMAP data and their respective clusters

```python
umap = pd.read_csv('../data/processed/primary_reduction_neighbors_15_components_2.csv')
```
