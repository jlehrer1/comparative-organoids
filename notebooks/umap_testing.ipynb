```python
import numpy as np 
import matplotlib.pyplot as plt 
import plotly.express as px 
import plotly.graph_objects as go
import umap
import pandas as pd 
```

```python
pts = np.array([np.random.rand(10000) for _ in range(1000)])
```

```python
res = umap.UMAP(
    n_neighbors=200, 
    min_dist=0.0001,
    n_components=3,
)
```

```python
for _ in range(10):
    pts = np.array([np.random.rand(1000) for _ in range(1000)])
    pts = pd.DataFrame(res.fit_transform(pts))
    fig = px.scatter_3d(pts, x=0, y=1, z=2)
    fig.update_traces(marker=dict(size=2))
#     fig.update_xaxes(range=[-10000, 10000])
#     fig.update_yaxes(range=[-10, 10])
    fig.show()
```
